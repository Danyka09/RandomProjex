```python
# slicing = create a substring by extracting elements from another string
#           indexing[] or slice()
#           [start:stop:step]

name = "Danny Boi"
```

```python
first_name = name[0:5]
print(first_name)
```

```python
last_name = name[6:9]
print(last_name)
```

```python
funky_name = name[::2]
print(funky_name)
```

```python
reversed_name = name[::-1]
print(reversed_name)
```

SLICE FUNCTION

```python
website1 = "http://google.com"
website2 = "http://wikipedia.com"

slice = slice(7,-4) # this is useful for repetition cause it creates a slice object we can reuse infinitely. but it works the same way
# start, end, step (but we use commas instead)

website1_name = website1[slice]
website2_name = website2[slice]
print(website1_name)
print(website2_name)
```
