```python
temperature = 75

if temperature < 32:
    print("Wear a heavy coat!")
elif temperature < 50:
    print("Wear a jacket!")  # elif only executes if "if" is false, so we dont need to check if its below 50 cause we already did
elif temperature < 65:
    print("Bring a light layer.")
elif temperature < 80:
    print("No jacket needed!")
else:
    print("Way too hot for a jacket!")

    #RULES
    # 1. Python evaluetes conditions from top to bottom
    # 2. Python executes the first statement that true or the else statement if no conditions are true
    # 3. Python skips all statements after getting one with a true condition
```

```python
temperature = 45
is_raining = True
# Logical expression
if temperature < 50 and is_raining: # "and" requires both conditions to be true' evaluetes to true if both sides are true
    print("Wear a warm rainjacket")
else:
    print("No jacket needed")
```
