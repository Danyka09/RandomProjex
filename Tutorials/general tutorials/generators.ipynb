```python
# https://youtu.be/GWZf_B129zs?si=HU5rUvcgDIeepNuv

# a generator function uses yield instead of return
# in a normal function we use return, get the value and python stops executing the function, if it contains more lines of code after return they won't be executed
# if we use yield, after getting the value python pauses the execution of the function, we can tell python to resume the execution of the code. this is useful when we have multiple yield statements, we can get the values one at a time
```

```python
def gen_values():
    yield 1
    yield 2
    yield 3

gen_values_obj = gen_values()
next(gen_values_obj)
# #when we only call the generator function it just creates a special object, we dont get a value, for that we use next() to resume code; the next() function tells the generator to run the function until it reaches the next yield
```

```python
def gen_values():
    yield 1
    yield 2
    yield 3

gen_values_obj = gen_values()

print(next(gen_values_obj))
print(next(gen_values_obj))
print(next(gen_values_obj))
```

```python
def gen_values():
    yield 1
    yield 2
    yield 3

gen_values_obj = gen_values()
# better way is to use a loop
for value in gen_values_obj:
    print(value)
```

```python
def gen_values():
    yield 1
    yield 2
    yield 3

# in the loop we dont even need a separate variable, we can dirctly call the function
for value in gen_values():
    print(value)
```

```python
def get_primes_list(start, end):
    primes = []
    for num in range(start, end + 1):
        if num < 2:
            continue
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
    return primes

get_primes_list(50, 100)
```

```python
def get_primes_generator(start, end):
    for num in range(start, end + 1):
        if num < 2:
            continue
        is_prime = True
        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            yield num

gen_obj = get_primes_generator(50, 100)
print(next(gen_obj))
print(next(gen_obj))
print(next(gen_obj))
```
