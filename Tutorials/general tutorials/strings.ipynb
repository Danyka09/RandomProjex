```python
result = name.upper() # capitalizes all letters of a string
```

```python
result = name.lower() # makes all letters of a string lowercase
```

```python
result = name.isdigit() # returns true if the string is only digits (numbers)
```

```python
result = name.isalpha() # returns true if the string contains only alphabetical characters (space is not included)
```

```python
result = name.count("y") # counts the number of occurrences of a given character
```
