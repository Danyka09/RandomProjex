```python
# kinda works but not the best way
multi_input = []
while True:
    input1 = input()
    if input1 != "":
        multi_input.append(input1)
    else:
        break

print(multi_input)
```

```python
import sys

# Prečíta úplne všetko až do konca súboru
data = sys.stdin.read()

# Ak to chceš rozbiť na riadky do zoznamu:
riadky = data.splitlines()

print(riadky)
```

```python
import sys

vsetko = []
for riadok in sys.stdin:
    vsetko.append(riadok.strip()) # .strip() odstráni \n na konci

print(vsetko)
```
