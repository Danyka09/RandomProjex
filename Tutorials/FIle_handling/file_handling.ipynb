```python
# the open() function takes the name of the file we want to work with and the mode (r for reading)
file = open("shopping_list.txt", "r")
contents = file.read() # .read() reads the file obviously.
file.close() # .close() closes the file, very important
print(contents) # even though its on different lines its the same string with hidden \n
print(repr(contents)) #prints raw representation
```

```python
file = open("shopping_list.txt")
contents = file.read()
file.close()

items = contents.split("\n")
items.sort()

for i in items:
    print("-" + i)
```

```python
#writing to a file
items = "eggs\nbread\nmilk"
file = open("groceries.txt", "w") # creates a new file called groceries; w is write but use with caution, can overwrite
file.write(items)
file.close()
```

```python
# writing to an existing file without overwriting
items = "\nyogurt" # watch out for new line otherwise it wont do it by itself
file = open("groceries.txt", "a") # a for append
file.write(items)
file.close()
```

```python
#context managers - we dont need to call close()
with open("groceries.txt", "a") as file:
    file.write("\nbananas")
```

```python
# this is how we specify a location, on windows an r before the string is necessary cause windows uses \ which messes with python
with open(r"C:\Users\dsebo\Desktop\groceries.txt", "a") as file: # this is an absolute file path
    file.write("apples")
```
